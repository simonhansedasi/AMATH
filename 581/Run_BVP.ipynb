Solve
$$y'' + (100-\beta)y + y^3 = 0$$
with b.c.s
$$y(1) = 0, y(-1) = 0, y'(-1) = A = 1$$

$$y_0 = y$$
$$y_1 = y'$$

```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
```

```python
# defining parameters
beta = 99;
x = np.linspace(-1,1,50)
```

```python
# defining functions
def rhs_bvp(x,y,beta):
    rhs = np.array([y[1],(beta - 100)*y[0] - 1000*y[0]**3])
    return rhs

def bvp_bc(yl,yr,beta):
    bc = np.array([yl[0],yl[1]-1,yr[0]])
    return bc

def init_sol(x):
    init = np.array([np.cos((np.pi/2)*x),(-np.pi/2)*np.sin((np.pi/2)*x)])
    return init
```

```python
sol = scipy.integrate.solve_bvp(rhs_bvp,bvp_bc,x,init_sol(x), p=[beta])
```
