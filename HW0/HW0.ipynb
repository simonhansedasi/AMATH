```python
# Question 3
x = [1, 3, 4, 8, 9]
y = [3, 4, 5, 7, 12]
z = np.polyfit(x, y, 1)

fig, ax = plt.subplots(1,1,figsize=(15, 12))
plt.subplots_adjust(hspace=0.5)
plt.suptitle('Practice Plot' + 
             '\nBest fit for given data',
             fontsize=18, y=0.95)
fig.patch.set_facecolor('w')

plt.scatter(x, y, label = 'Given data')

x_line_fit = np.arange(1, 10)
y_line_fit = np.polyval(z, x_line_fit)

ax.plot(x_line_fit, y_line_fit, label='Line of best fit')

from scipy.stats import linregress
r = linregress(x_line_fit,y_line_fit)
print(r)                                                             
plt.plot(r, label = 'Scipy linear regression')
ax.legend(loc='upper left', fontsize=13 )

slope = np.polyfit(x_line_fit,y_line_fit, 1)[0]
print(slope)
```
